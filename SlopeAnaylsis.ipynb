```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x = []
y = []
z =np.loadtxt("./3.csv",delimiter= ',')
csvx = z[0][1:]
csvy = z[1][1:]
csvx.shape , csvy.shape
```

```python
slopes = np.array([])
for i in range(len(x) -1):
    slope = (y[i+1] - y[i])/(x[i+1] - x[i])
    slopes = np.append(slopes, slope)
```

```python
slopes2d = np.array([])
for i in range(len(x) -2):
    slope2d = (slopes[i+1] - slopes[i])/(x[i+1] - x[i])
    slopes2d = np.append(slopes2d, slope2d)
slopes2d
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111)

x, y = [], []

def animate(i):
    if i < len(csvx) and i < len(csvy):
        x.append(csvx[i])
        y.append(csvy[i])
        ax.plot(x, y, color="blue")
  


ani = FuncAnimation(fig=fig, func=animate, interval=300).save("animation.gif")

```
