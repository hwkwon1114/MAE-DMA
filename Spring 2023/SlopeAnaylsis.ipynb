```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x = []
y = []
z =np.loadtxt("./Last.csv",delimiter= ',')
csvx = z[1][1:]
csvy = z[0][1:]
csvx.shape , csvy.shape
```

Position Vs Slope

```python
slopes = np.array([])
for i in range(len(csvx) -1):
    slope = (csvy[i+1] - csvy[i])/(csvx[i+1] - csvx[i])
    slopes = np.append(slopes, slope)
fig, ax1 = plt.subplots()
ax1.plot(csvx[-30:] , slopes[-30:])
poly = np.polyfit(csvx[-30:], slopes[-30:], 2)
z = np.poly1d(poly)
ax1.plot(csvx[-30:],z(csvx[-30:]))
ax1.set_xlabel("Position(mm)")
ax1.set_ylabel("Slope(V/mm)")
ax1.set_title("Determining Change in Slope")
```

```python
slopes2d = np.array([])
for i in range(len(csvx) -2):
    slope2d = (slopes[i+1] - slopes[i])/(csvx[i+1] - csvx[i])
    slopes2d = np.append(slopes2d, slope2d)
slopes.shape
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111)
ax2 = fig.add_subplot(111)
linx = np.linspace(1, len(csvx),len(csvx))
slopes = np.array([])
slopes = np.append(slopes, 0)
for i in range(len(csvx) -1):
    slope = (csvy[i+1] - csvy[i])/(csvx[i+1] - csvx[i])
    slopes = np.append(slopes, slope)
x, y , s= [], [], []
ax.set_xlabel('Data Index')
ax.set_ylabel('Amplitude(V)')
ax2.set_ylabel('Slope(V/mm)')
ax.set_title('Amplitude and Slope vs Position')
ax.set_label("Amplitude(V)")
ax2.set_label("Slope(V/mm)")
amplitude, = ax.plot([], [], color="blue", label='Amplitude(mm)')
slope, = ax2.plot([], [], color="red", label='Slope(V/mm)')
ax.legend(handles=[amplitude, slope], loc='upper left')
def animate(i):
    if i < len(csvx) and i < len(csvy):
        x.append(linx[i])
        y.append(csvy[i])
        s.append(slopes[i])
        ax.plot(x,y, color="blue")
        ax2.plot(x,s,color = "red")
        if (i == len(csvx) - 1):
            ax2.plot(x[-30:],z(csvx[-30:]))
ani = FuncAnimation(fig=fig, func=animate,interval=400, repeat_delay = 100).save("animation5.gif")
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(9, 6))
fig.subplots_adjust(hspace=0.3)
linx = np.linspace(1, len(csvx), len(csvx))
slopes = np.array([])
slopes = np.append(slopes, 0)
for i in range(len(csvx) - 1):
    slope = (csvy[i + 1] - csvy[i]) / (csvx[i + 1] - csvx[i])
    slopes = np.append(slopes, slope)

x1, y1 ,s = [], [],[]

ax1.set_xlabel('Data Index')
ax1.set_ylabel('Amplitude(V)')
ax2.set_ylabel('Slope(V/mm)')
ax2.set_xlabel('Data Index')
ax1.set_title('Amplitude and Slope vs Index')

def animate(i):
    if i < len(csvy):
        x1.append(linx[i])
        y1.append(csvy[i])
        s.append(slopes[i])
        ax1.plot(x1, y1, '-o',color = "blue")
        ax2.plot(x1, s, '-o',color = "red")
        
        if i == len(csvx) - 1:
            ax2.plot(x1[-30:], z(csvx[-30:]), color = "blue")

ani = FuncAnimation(fig=fig, func=animate, interval=400, repeat_delay=400).save("animation5.gif")
```
